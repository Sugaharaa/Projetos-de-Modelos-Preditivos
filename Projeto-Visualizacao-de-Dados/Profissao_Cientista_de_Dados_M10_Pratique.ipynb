```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv('ecommerce_preparados.csv')
print(df.head())

#Histograma
plt.hist(df['Nota'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribuição das Notas')
plt.xlabel('Nota')
plt.ylabel('Quantidade')
plt.show()

# Gráfico de Dispersão
plt.figure(figsize=(6, 4))
plt.scatter(df['Nota'], df['N_Avaliações'], color='blue', label='Nota x Numero de avaliações')
plt.title('Gráfico de Dispersão - Matplotlib')
plt.xlabel('Nota')
plt.ylabel('Numero de avaliações')
plt.legend()
plt.show()

# Mapa de Calor
plt.figure(figsize=(10, 8))
correlacao = df.select_dtypes(include=['number']).corr()
sns.heatmap(correlacao, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Mapa de Calor - Correlação entre Variáveis')
plt.show()

# Contagem de avaliações por Nota
nota_counts = df['Nota'].value_counts()

# Gráfico de Barra
plt.figure(figsize=(8, 6))
nota_counts.sort_index().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Número de Avaliações por Nota')
plt.xlabel('Nota')
plt.ylabel('Quantidade de Avaliações')
plt.xticks(rotation=0)
plt.show()

# Gráfico de Pizza
plt.figure(figsize=(8, 8))
plt.pie(nota_counts, labels=nota_counts.index, autopct='%1.1f%%', colors=sns.color_palette('Blues', len(nota_counts)), startangle=90)
plt.title('Distribuição das Notas em Percentual')
plt.show()

# Gráfico de Densidade
plt.figure(figsize=(8, 6))
sns.kdeplot(df['Nota'], shade=True, color='skyblue', alpha=0.7)
plt.title('Distribuição de Densidade das Notas')
plt.xlabel('Nota')
plt.ylabel('Densidade')
plt.show()

# Gráfico de Regressão
plt.figure(figsize=(8, 6))
sns.regplot(x='Nota', y='N_Avaliações', data=df, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
plt.title('Gráfico de Regressão: Nota vs Número de Avaliações')
plt.xlabel('Nota')
plt.ylabel('Número de Avaliações')
plt.show()
```
